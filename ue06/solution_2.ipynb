### Exercise 2
A *break point* of hypothesis class $\mathcal{H}$ is the smallest $n$ such that $m_{\mathcal{H}}(n) < 2^n$

**(a)** Consider the hypothesis classes *decision point*, *interval*, and *convex set* defined in vl05, page 39ff. Find a break point, if any, for each hypothesis class. Plot the growth functions for each hypothesis class (use log scale for the y-axis). 

**(b)** By inspection, find a break point for the hypothesis class of axis-aligned rectangles in $\R^2$. 


---
### Solution to (a)

**Decision Point:** The break point is $2$, because  

$$
m_{\mathcal{H}}(n) = n+1 = 2^n 
$$

holds for $n = 1$ and is violated for $n\geq 2$. By inspection, the hypothesis space can shatter only a single point. For two points $x_1 < x_2$, there is no decison point classifier $f$ that can realize the labeling 

1. $f(x_1) = +1$
2. $f(x_2) = -1$


**Interval:** The break point is $3$, because  

$$
m_{\mathcal{H}}(n) 
= 1+\binom{n+1}{2} 
= \frac{1}{2}n^2 + \frac{1}{2}n + 1
= 2^n 
$$

holds for $n \leq 2$ and is violated for $n\geq 3$. By inspection, the hypothesis space can shatter at most two points, but fails to shatter three points. For three points $x_1 < x_2 < x_3$, there is no interval classifier $f$ that can realize the labeling 

1. $f(x_1) = +1$
2. $f(x_2) = -1$
3. $f(x_3) = +1$

**Convex set:** There is no break point, because  

$$
m_{\mathcal{H}}(n) = 2^n 
$$

for all $n$. 

**Plot:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
n = np.arange(1, 10)
m1 = n+1
m2 = comb(n+1, 2)+1
m3 = 2**n

plt.semilogy(n, m1, lw=2, label='decision')
plt.semilogy(n, m2, lw=2, label='interval')
plt.semilogy(n, m3, lw=2, label='convex')
plt.xlabel('$x$')
plt.ylabel('$m_{\mathcal{H}}(n)$')
plt.legend()
plt.show()
```

### Solution to (b)

The break point of axis-aligned rectangles in the plane os five. 

An axis-aligned hypothesis space can shatter four points in a diamond pattern, realizing all 16 labelings.

However, it cannot shatter any set of five distinct points in the plane. Label the following points as positive:

1. the point with smallest x-coordinate
2. the point with largest x-coordinate
3. the point with smallest y-coordinate
4. the point with largest y-coordinate

Since the points are distinct, there are at least two points that satisfy all four conditions. The smallest axis-aligned rectangle containing these points includes all remaining points. If one of the remaining points is labeled as negative, no axis-aligned rectangle can realize this labeling.
