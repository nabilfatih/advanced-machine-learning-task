### Exercise 4

Plot the VC bounds as a function of the sample size $n$ and as function of the error tolerance $\varepsilon$ for VC dimensions $1$, $3$ and $5$. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

---
### Computing the VC Bound

```python
def vc_bound(d, n, eps):
    sauer = (2*np.e*n/d)**d
    return 4*sauer*np.exp(-n*eps**2/8)
```

---
### VC Bound as Function of n

```python
n = np.arange(1, 5001, 10)
vcdim = [1, 3, 5]
eps = 0.2
```

```python

for d in vcdim:
    b = vc_bound(d, n, eps)
    plt.loglog(n, b, lw=2, label=f'vc-dim = {d}')
plt.xlabel('n')
plt.title('VC Bounds')
plt.legend()
plt.tight_layout()
plt.show()
```

---
### VC Bound as Function of eps

```python
n = 1000
vcdim = [1, 3, 5]
eps = np.arange(0.01, 1.01, 0.01)
```

```python
for d in vcdim:
    b = vc_bound(d, n, eps)
    plt.semilogy(eps, b, lw=2, label=f'vc-dim = {d}')
plt.xlabel('$\epsilon$')
plt.title('VC Bounds')
plt.legend()
plt.tight_layout()
plt.show()
```
