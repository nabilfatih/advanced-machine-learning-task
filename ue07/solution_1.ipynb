### Notation 

1. Let $\langle x, x'\rangle$ denote the scalar product of points $x, x' \in \mathbb{R}^d$. 

2. The sign function is defined as

$$
\text{sgn}:\mathbb{R} \to \mathbb{R}, \quad u \mapsto 
\begin{cases}
+1 & u \geq 0\\
-1 & u < 0
\end{cases}.
$$

---
### Exercise 1

The hypothosis class of homogeneous halfspaces (linear classifiers without bias) in $\mathbb{R}^d$ is defined by 

$$
\mathcal{H}=\{f:\mathbb{R}^d\to \{\pm 1\} \;:\; \exists\, w\in\mathbb{R}^d \text{\ s.t.\ } f(x)=\text{sgn}\big(\langle w, x\rangle\big)\}.
$$

Show that the VC dimension of $\mathcal{H}$ is $d$.


**Hints:** 
First show (a) that $\mathcal{H}$ shatters $d$ points in $\mathbb{R}^d$. Next show (b) that $\mathcal{H}$ cannot shatter any set of $d+1$ points in $\mathbb{R}^d$. Exploit linear independence in (a) and linear dependence in (b).

---
**Proof (a):**

Let $\mathcal{X}_d = \{e_1, \ldots, e_d\}$ be the standard basis vectors of $\mathbb{R}^d$. This set $\mathcal{X}_d$ is shattered by $\mathcal{H}$: For each labeling $y_1, \ldots, y_d$ of $\mathcal{X}_d$, set $w = (y_1, \ldots, y_d)$. Then $f(e_i) = \langle w, e_i \rangle = y_i$ for all $i$.

---
**Proof (b):**

Assume that $\mathcal{H}$ shatters the set $\mathcal{X}_{d+1} = \{x_1, \ldots, x_{d+1}\} \subseteq \mathbb{R}^{d}$. 

Note that $0\notin\mathcal{X}_{d+1}$, because $\langle 0, w \rangle = 0$ for all $w\in \mathbb{R}^d$. Thus, $0$ cannot be shattered. Since $\mathcal{X}_{d+1}$ is linear dependent, there exist numbers $a_1, \ldots, a_{d+1} \in \mathbb{R}$ with at least one $a_k\neq 0$ such that 

$$
a_1 x_1 + \cdots a_{d+1}x_{d+1} = 0.
$$

Let $\mathcal{I} = \{i \,:\, a_i > 0\}$ and let $\mathcal{J} = \{j \,:\, a_j < 0\}$. We distinguish between two cases:

1. $\mathcal{I}\neq \emptyset$ and $\mathcal{J}\neq \emptyset$ 
2. $\mathcal{I} = \emptyset$ or $\mathcal{J} = \emptyset$ 

Note that by linear dependence, at least one of both sets $\mathcal{I}$ and $\mathcal{J}$ is non-empty. 

**Case 1:** If both $\mathcal{I}$ and $\mathcal{J}$ are non-empty, then

$$
\sum_{i \in \mathcal{I}} a_i x_i = \sum_{j \in \mathcal{J}} |a_j| x_j.
\tag{1}
$$

Consider the labeling 

$$
y_k = \begin{cases}
+1 & k \in \mathcal{I}\\
-1 & k \in \mathcal{J}
\end{cases}
$$

for all $k\in\{1, \ldots, d+1\}$. Since $\mathcal{H}$ shatters $\mathcal{X}_{d+1}$, there exists a weight vector $w\in \mathbb{R}^d$ such that $\langle x_i, w \rangle > 0$ for all $i \in \mathcal{I}$ and $\langle x_j, w \rangle < 0$ for all $j \in \mathcal{J}$ (see note at the end of the proof). 

Using Equation (1), we obtain a contradiction:

$$
0 < \sum_{i \in \mathcal{I}} a_i \langle x_i, w \rangle
= \left\langle\sum_{i \in \mathcal{I}} a_i x_i, w \right\rangle
= \left\langle\sum_{j \in \mathcal{J}} |a_j| x_j, w \right\rangle
= \sum_{j \in \mathcal{J}} |a_j| \langle x_j, w \rangle < 0.
$$


**Case 2:** Suppoose that either $\mathcal{I} = \emptyset$ or $\mathcal{J} = \emptyset$. We show the case for $\mathcal{I}= \emptyset$. The case $\mathcal{J} = \emptyset$ follows a similar argument. 

From $\mathcal{I} = \emptyset$ follows $\sum_{j \in \mathcal{J}} a_j x_j = 0.$ Similarly to the first case, we obtain a contradiction: 

$$
0 = \left\langle\sum_{j \in \mathcal{J}} a_j x_j, w \right\rangle
= \sum_{j \in \mathcal{J}} a_j \langle x_j, w \rangle
> 0.
$$

Combining both cases, we conclude that $\mathcal{H}$ cannot shatter $d+1$ points. 

**Note:** 
Suppose $\mathcal{H}$ shatters $\mathcal{X}_{d+1}$. In the proof, we asserted the existence of a weight vector $w\in \mathbb{R}^d$ such that $\langle x_i, w \rangle > 0$ for all $i \in \mathcal{I}$ and $\langle x_j, w \rangle < 0$ for all $j \in \mathcal{J}$. 

However, for $\mathcal{H}$ to shatter $\mathcal{X}_{d+1}$, a strict inequality for positive examples is not necessary. Therefore, all we can guarantee is the existence of a weight vector $w'\in \mathbb{R}^d$ such that $\langle x_i, w' \rangle \geq 0$ for all $i \in \mathcal{I}$ and $\langle x_j, w' \rangle < 0$ for all $j \in \mathcal{J}$.

To obtain the desired weight vector $w\in \mathbb{R}^d$, first recall that $0$ cannot be an element of $\mathcal{X}_{d+1}$. Next, suppose that
$\mathcal{K} \subseteq \mathcal{I}$ is the subset of indices such that  $\langle w, x_k\rangle = 0$ for all $k \in \mathcal{K}$. 

Pick $k \in \mathcal{K}$ and define $w = w' + \varepsilon x_k$ for a sufficiently small $\varepsilon > 0$. 

Then, $\langle w, x_k\rangle > 0$, because

$$
\begin{align*}
\langle w, x_k\rangle &> 0 \tag{1}\\ 
\langle w, x_i\rangle &> 0 
    \quad \forall i \in \mathcal{I}\setminus\mathcal{K}\tag{2}\\
\langle w, x_j\rangle &< 0 
    \quad \forall j \in \mathcal{J}\tag{3}
\end{align*}
$$

Equation (1) follows from
$$
\langle w, x_k\rangle 
= \langle w'+\varepsilon x_k, x_k\rangle 
= \langle w', x_k \rangle + \varepsilon \langle x_k, x_k\rangle
= \varepsilon \langle x_k, x_k\rangle > 0.
$$

Equation (2) follows from 
$$
\langle w, x_i\rangle 
= \langle w'+\varepsilon x_k, x_i\rangle 
= \langle w', x_i \rangle + \varepsilon \langle x_k, x_i\rangle > 0,
$$
for all $i \in \mathcal{I}\setminus\mathcal{K}$, because $\varepsilon$ is sufficiently small. With the same argument, Equation (3) follows from 

$$
\langle w, x_j\rangle 
= \langle w'+\varepsilon x_k, x_j\rangle 
= \langle w', x_j \rangle + \varepsilon \langle x_k, x_j\rangle < 0
$$

for all $j \in \mathcal{J}$. 

Remove $k$ from $\mathcal{K}$ and add it to $\mathcal{I}$. Thus, we reduced $\mathcal{K}$ by one element. Repeat the same process until $\mathcal{K} = \emptyset$ to obtain the desired weight vector.
