### Exercise 0

---
**Karl's Conjecture:** Let $x_1, \ldots, x_n \in [a, b]$ be i.i.d. random variables with mean $\mu=\mathbb{E}[x_i]$. Let 

$$
z = \sum_{i=1}^n (x_i -\mu).
$$

Then 

1. $\mathbb{E}[z] = 0$

2. For all $s>0$, we have
    $$
    \mathbb{E}[\exp(sz)] \leq \exp\left(\frac{n^2}{8}s^2(b-a)^2\right)
    $$

---
**Task:** Prove the second statement directly, without decomposing $z$ into individual random variables $z_i = x_i - \mu$ as in Step 2 of the proof of Hoeffding's inequality. There is no constraint on the proof of item (1).

Do we need independence of the random variables $x_1, \ldots, x_n$? What is the price we pay for directly proving the second statement?

---
**Proof:**

Part 1: Since $z=\sum_i (x_i - \mu)$, we have

$$
\mathbb{E}[z] = \mathbb{E}[\sum_i (x_i - \mu)] = \sum_i (\mathbb{E}[x_i] - \mathbb{E}[\mu])= \sum_i (\mu-\mu) = 0.
$$

Part 2: To apply the Lemma from Hoeffding's proof, we need to show that $z$ is bounded with expectation zero. 

1. *$z$ is bounded:* From $a \leq x_i \leq b$ follows

    $$
    \sum_i (a-\mu) \leq \sum_i (x_i - \mu) \leq \sum_i (b - \mu).
    $$

    Thus, $a'=n(a-\mu)\leq z\leq n(b-\mu)=b'$.

2. *$\mathbb{E}[z] = 0$:* Shown in the first part of this proof. 

Note that $b'-a' = n(b-a)$. Applying the Lemma of Hoeffding's proof gives

$$
\begin{align*}
\mathbf{E}[\exp(sz)] 
&\leq \exp\left(\frac{1}{8}s^2(b'-a')^2\right)\\
&= \exp\left(\frac{n^2}{8}s^2(b-a)^2\right).
\end{align*}
$$

**Note:** The proof does not require that the random variables $x_1, \ldots, x_n$ are independent. The bound for directly applying the Lemma of Hoeffding's proof is not as tight as in Hoeffding's proof:

$$
\mathbf{E}[\exp(sz)] 
\leq \exp\left(\frac{n}{8}s^2(b-a)^2\right).
$$
