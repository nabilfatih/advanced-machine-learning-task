### Exercise 3

Prove the following statement:

> AdaBoostâ€™s optimal decision function is of the form
> 
> $$
f^* (x) = \argmin_{f \in \mathcal{L}(\mathcal{B},T)} \mathbb{E}_{y|x} [\exp(-yf(x))] = \frac{1}{2} \ln \left( \frac{\mathbb{P}(y=+1 | x)}{\mathbb{P}(y=-1 | x)} \right)
> $$
> or equivalently 
> $$
\mathbb{P}(y=+1 | x) = \frac{1}{1+\exp(-2f^* (x))}.
>$$

What does this statement remind you of?

---

The term 

$$
\frac{1}{2} \ln \left( \frac{\mathbb{P}(y=+1 | x)}{\mathbb{P}(y=-1 | x)} \right)
$$

is half of the log-odds ratio and the term 

$$
\frac{1}{1+\exp(-2f^* (x))}
$$

is the sigmoid with factor $2$. Thus, as logistic regression, AdaBoost models the posterior.

In **logistic regression**, the log-odds of the probability of a positive outcome is modeled as a linear combination of the predictor variables. This is represented by the logit function:

$$
\ln \left( \frac{p(x)}{1 - p(x)} \right) = w_0 + w_1x_1 + \cdots + w_nx_n.
$$

The optimal decision function for **AdaBoost** is derived as:

$$
f^*(x) = \frac{1}{2} \ln \left( \frac{p(x)}{1 - p(x)} \right)
$$

or equivalently,

$$
p(x) = \frac{1}{1+\exp(-2f^*(x))}
$$

which is similar to the form of the logistic regression model.

The similarity arises because both methods are trying to model the same thing: the conditional probability of the class $y$ given the input $x$. However, they approach the problem in different ways. Logistic regression assumes that the log-odds is linear, while AdaBoost makes no such assumption and instead learns the conditional probability from the data sequentially in several rounds in an adaptive and iterative manner. 

---
**Proof:**

Let $p(x) = \mathbb{P}(y=+1 | x)$. Then we have

\begin{align*}
\mathbb{E}_{y|x} [\exp(-yf(x))] &= p(x) \cdot \exp(-f(x)) + (1 - p(x)) \cdot \exp(f(x)) \\
&= p(x) \cdot \exp(-z) + (1 - p(x)) \cdot \exp(z) \\
&= g(z)
\end{align*}

where $z = f(x)$. We minimize $g(z)$ by equating the derivative $g'(z)$ to zero and solving for $z$:

\begin{align}
g'(z) &= -p(x) \cdot \exp(-z) + (1 - p(x)) \cdot \exp(z) = 0
\end{align}

Solving for $z$ yields

\begin{align*}
&{\quad} 0 = -p(x) \cdot \exp(-z) + (1 - p(x)) \cdot \exp(z) \\[2ex]
\Leftrightarrow&{\quad} p(x) \cdot \exp(-z) = (1 - p(x)) \cdot \exp(z) \\[2ex]
\Leftrightarrow&{\quad} \frac{p(x)}{1 - p(x)} = \exp(2z) \\[2ex]
\Leftrightarrow&{\quad} \ln \left( \frac{p(x)}{1 - p(x)} \right) = 2z \\[2ex]
\Leftrightarrow&{\quad} z = \frac{1}{2} \ln \left( \frac{p(x)}{1 - p(x)} \right)
\end{align*}

This proves
\begin{align*}
f^*(x) = z^* = \frac{1}{2} \ln \left( \frac{p(x)}{1 - p(x)} \right).
\end{align*}

We show the equivalent formulation
$$
p(x) = \frac{1}{1+\exp(-2f^*(x))}.
$$

by solving Equation (1) for $p(x)$:

\begin{align*}
&{\quad} 0 = -p(x) \cdot \exp(-z) + (1 - p(x)) \cdot \exp(z) \\[2ex]
\Leftrightarrow&{\quad} p(x) \cdot (\exp(-z) + \exp(z)) = \exp(z) \\[2ex]
\Leftrightarrow&{\quad} p(x) = \frac{\exp(z)}{\exp(-z) + \exp(z)} \\[2ex]
\end{align*}

Observe that 

$$
\frac{\exp(z)}{\exp(-z) + \exp(z)} = \frac{\exp(z)}{\exp(-z) + \exp(z)} \frac{\exp(-z)}{\exp(-z)} = \frac{1}{1+\exp(-2z)}
$$

Substituting $f^*(x)$ for $z$ yields the assertion.
